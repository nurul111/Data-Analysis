```python
df2 = pd.read_csv("Assessment sheet - Data 2.csv")
df2.tail(2)
```

```python
#Q1.Find courses that contain the word "course"  in it's courseTitle
df =df1[df1['courseTitle'].str.contains('Course',case=False)]
print("Courses that contain the word 'course' in it's courseTitle:\n")
df.head(2)
```

```python
#Q2.courseProvider, offerPrice and unitSold for Data 2
print("courseProvider, offerPrice and unitSold for Data 2: \n")
df = pd.merge(df1[["courseId","courseProvider","offerPrice","unitSold"]], df2, how="inner", on="courseId")
df
```

```python
url = []
for i in df2["courseId"]:
   url.append(f"https://www.reed.co.uk/courses/demo/{i}")
df2["url"]=url

df2
```

```python
#Q3.Make courseLink for Data 2(All)

def make_clickable(url,name):
    return '<a href="{}" rel="noopener noreferrer" target="_blank">{}</a>'.format(url,name)

df2['link'] = df2.apply(lambda x: make_clickable(x['url'],x['courseId']), axis=1)
df2.style
```

```python
#Q4.Find courses which unitSold are equal or below 5000 and offerPrice are above 50
filt = (df1["unitSold"]<=5000) & (df1["offerPrice"]>50)
df1.loc[filt,["courseId","courseTitle","unitSold","offerPrice"]]
```

```python
#Q5.Find 10 top sold courses
filt = df1.sort_values(by=['unitSold'],ascending=False)
filt[["courseId","courseTitle","unitSold"]].head(10)
```

```python
#Q6 I want to change all the Training Express(A provider name) into Rawalpindi Express. How to do that?
filt = df1["courseProvider"].str.contains('Training Express',case=False)
df1.loc[filt,"courseProvider"]="Rawalpindi Express"
df1.head(72)
```

```python
#Q7.Make the whole cell red when the courseProvider is "Learning Curve Group"
def color_red(val):
    color = 'red' if val == "Learning Curve Group" else ''
    return 'background-color: %s' % color
df1.style.applymap(color_red, subset='courseProvider')
```
