```python
import datetime as DT
import numpy as np
```

```python
data['dob'] = pd.to_datetime(data.dob)
data.style.format({"dob": lambda t: t.strftime("%d-%m-%Y")})
data.head(20)
```

```python
now = pd.Timestamp('now')
data['dob'] = pd.to_datetime(data['dob'], format='%d-%m-%Y')    # 1
data['dob'] = data['dob'].where(data['dob'] < now, data['dob'] - np.timedelta64(100, 'Y'))   # 2
data['age'] = (now - data['dob']).astype('<m8[Y]')   # 3
data.head(20)
```

```python
data.shape
```

```python
print(data['income'].max())
```

```python
ig = []
for i in data['income']:
    if i<=33000:
        ig.append('Low')
    elif i>33000 and i<=66000:
        ig.append('Medium')
    elif i>66000:
        ig.append('High')
```

```python
print(len(ig))
```

```python
data['income_group'] = ig 
```

```python
data.head(20)
```

```python
print(data['score'].max())
```

```python
sg = []
for i in data['score']:
    if i<=300:
        sg.append('Poor')
    elif i>300 and i<=600:
        sg.append('Fair')
    elif i>600:
        sg.append('Good')
```

```python
print(len(sg))
```

```python
data['score_group'] = sg
```

```python
data.head(20)
```

```python
data['last_purchase_date'] = pd.to_datetime(data.last_purchase_date)
data.style.format({"last_purchase_date": lambda t: t.strftime("%Y-%m-%d")})
data.head(20)
```

```python
lpd = data[ (data['last_purchase_date'] < '2019')].index
data.drop(lpd , inplace=True)
```
